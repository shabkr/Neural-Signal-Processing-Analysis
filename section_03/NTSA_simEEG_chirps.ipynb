**Goal:** Simulate different types of chirps

```python
# Remember to always import first!

import matplotlib.pyplot as plt
import numpy as np
```

```python
# simulation details
pnts = 10000
srate = 1024
time = np.linspace(0,pnts,num=pnts)/srate
```

## Chirps

```python
# bipolar chirp
freqmod = np.linspace(5,15,num=pnts) #freqmod is the 'f' variable from the lecture

# multipolar chirp 
rng = np.random.default_rng() #random number generator, use see if you want the result to be the same every time
k = 10 #poles for frequencies
freqmod = 20*np.interp(x=time, xp=np.linspace(0,pnts/srate,num=k), fp=rng.random(k))

#signal time series
signal = np.sin(np.multiply(2*np.pi,(time + np.cumsum(freqmod)/srate)))
```

Now, we can plot the signal

```python
fig = plt.figure(layout="constrained")
ax_dict = fig.subplot_mosaic(
    [
        ["freq"],
        ["signal"]
    ]
)

ax_dict["freq"].plot(time,freqmod, "r-")
ax_dict["freq"].set_xlabel("Time (s)")
ax_dict["freq"].set_ylabel("Frequency (Hz)")
ax_dict["freq"].set_title("Instantaneous Frequency")

ax_dict["signal"].plot(time,signal)
ax_dict["signal"].set_xlabel("Time (s)")
ax_dict["signal"].set_ylabel("Amplitude (arbitrary units)")
ax_dict["signal"].set_title("Signal (chirp)")
```
